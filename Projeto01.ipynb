<h1>Breve Introdução

As Olímpiadas tiveram início na Grégia Antiga no século VIII a.C., e possuem este nome pois eram realizadas na cidade de Olímpia. Com o declínio da civilização grega, as competições esportivas se tornaram esparsas, tendo seu retorno no século XIX, com o objetivo de ser um evento para a celebração da paz. 

Fato interessante: Os Jogos de Inverno foram criados em 1924, e eram realizados no mesmo país e no mesmo ano das Olímpiadas de Verão. Em 1940 e 1944 os Jogos Olímpicos de Inverno foram cancelados por causa da Segunda Guerra Mundial, a partir daí sendo realizados em países diferentes do das Olímpiadas de Verão, mas ainda no mesmo ano. Até que em 1986, o COI (Cômite Olímpico Internacional) decidiu intercalar os Jogos de Verão e de Inverno.

O conjunto de dados (Dataset) nos fornece informações sobre os atletas e seus resultados ao longo dos anos, segue abaixo breve explicação sobre as colunas existentes e o que elas representam:

<b>ID / ID</b> - <I>Numeração que foi dada a cada evento, cada linha do dataset representa um evento, que é uma prova que o atleta participou.

<b>AGE / IDADE</b> - <I>Idade do atleta</I>

<b>NAME / NOME</b> - <I>Nome do atleta que competiu no evento</I>

<b>SEX / GÊNERO</b> - <I>Sexo do atleta participante (M = masculino, F = feminino)</I>

<b>HEIGHT / ALTURA</b> - <I>Altura em centímetros do atleta</I>

<b>WEIGHT / PESO</b> - <I>Peso em Quilogramas do atleta </I>

<b>TEAM / PAÍS</b> - <I>País que o atleta representa</I>

<b>NOC / SIGLA</b> - <I>Sigla do País do atleta</I>

<b>GAMES / JOGOS</b> - <I>Diz o ano e a temporada do evento que o atleta participou</I>

<b>YEAR / ANO</b> - <I>Ano do evento</I>

<b>SEASON / TEMPORADA</b> - <I>Diz se o evento foi realizado nos Jogos de Inverno ou de Verão</I>

<b>CITY / CIDADE</b> - <I>Cidade em que foi realizado o evento</I>

<b>SPORT / ESPORTE</b> - <I>Esporte que o evento representa</I>

<b>MEDAL / MEDALHA</b> - <I>Medalha que o atleta ganhou na prova </I

<h1>Tratando valores nulos</h1>

Um problema muito comum que encontramos ao trabalhar com grandes conjuntos de dados é a existência de valores nulos ou não preenchidos, que podem existir por erros de preenchimento, erros na importação e transformação dos dados, não existência da informação, ou qualquer outro motivo.

O tratamento que iremos dar para os valores nulos depende do nosso objetivo com a análise de dados, podemos deixar os valores nulos caso eles não impactem negativamente na nossa análise, podemos apagar a linha inteira que possuí um valor nulo, podemos preencher os valores nulos com um valor específico, existem inúmeras formas de tratar valores nulos, mas sempre devemos manter em mente o nosso objetivo com a análise de dados para decidir qual tratativa utilizar. 

Para o nosso trabalho iremos substituir os valores nulos pela média correspondente de cada coluna.

```python
#Visualizando a quantidade de valores nulos em cada coluna
df.isnull().sum()
```

```python
#Realizando a substituição dos valores nulos pela média de cada coluna
df['Idade'] = df['Idade'].fillna(df['Idade'].mean())
df['Altura'] = df['Altura'].fillna(df['Altura'].mean())
df['Peso'] = df['Peso'].fillna(df['Peso'].mean())
```

Verificando se nosso comando funcionou corretamente:

```python
df.isnull().sum()
```

Podemos ver que não existem mais valores nulos nas colunas "Idade", "Altura" e "Peso". A coluna medalha possuí valores nulos, mas não será relevante para nossa análise, então podemos deixar os valores dessa coluna na forma original.

<b>Quantidade de Atletas ao Longo do Tempo

```python
verao = df[df["Temporada"] == "Summer"]
inverno = df[df["Temporada"] == "Winter"]
```

```python
verao = verao.groupby(["Ano"])["ID"].nunique().reset_index()
inverno = inverno.groupby(["Ano"])["ID"].nunique().reset_index()

fig = plt.figure(figsize=(15,12))
plt.subplot(212)
ax = sns.pointplot(x = verao["Ano"] , y = verao["ID"],color="red",markers="h")
ax.set_facecolor("White")
plt.grid(True,alpha=.2, color="Black")
plt.ylabel("Contagem")
plt.title("Olímpiadas de Verão - Numero de atletas ao longo dos anos",color="black")

fig = plt.figure(figsize=(15,12))
plt.subplot(212)
ax = sns.pointplot(x = inverno["Ano"] , y = inverno["ID"],color="blue",markers="h")
ax.set_facecolor("White")
plt.grid(True,alpha=.2, color="Black")
plt.ylabel("Contagem")
plt.title("Olímpiadas de Inverno - Numero de atletas ao longo dos anos",color="black")
```

Podemos verificar a quantidade de atletas participantes dos Jogos Olímpicos de Verão e de Inverno ao longo dos anos em 2 gráficos separados.

Fato interessante: Em 1980 os EUA declaram o boicote às Olímpiadas de Verão realizadas em Moscou como represália à intervenção que a URSS estava realizando no Afeganistão. É possível visualizarmos uma redução no número de participantes no ano de 1980 no primeiro gráfico.

<h1>Medindo a Relação entre Variáveis</b>

<img src="img01.jpg",width=300, height=350>

Muitas vezes quando trabalhamos com análise de dados é de grande utilidade medirmos a relação entre duas variáveis, este conceito serve de base para criação de alguns tipos de modelos preditivos, explicando de forma simplificada: se sabemos a relação entre a variável <i>X</i> e a variável <i>Y</i> conseguimos criar um modelo matemático que possuindo apenas novas variáveis de entrada <i>X</i> irá prever a variável de saída <i>Y</i>.

A medida da força da correlação é expressa como um coeficiente, chamado de coeficiente de correlação de Pearson e ele assume valores entre -1 e 1.

- 0.9 a 1.0 positivo ou negativo indica uma correlação muito forte.
- 0.7 a 0.9 positivo ou negativo indica uma correlação forte.
- 0.5 a 0.7 positivo ou negativo indica uma correlação moderada.
- 0.3 a 0.5 positivo ou negativo indica uma correlação fraca.
- 0 a 0.3 positivo ou negativo indica uma correlação desprezível.

<b>Vamos agora explorar a relação entre as variáveis do nosso conjunto de dados:

```python
#gerando uma tabela com a correlação entre todas as variáveis do nosso dataset
df.corr()
```

Outra forma visual de verificarmos a correlação entre as variáveis:

```python
def plot_corr(df, size=8):
    corr = df.corr()    
    fig, ax = plt.subplots(figsize = (size, size))
    ax.matshow(corr)  
    plt.xticks(range(len(corr.columns)), corr.columns) 
    plt.yticks(range(len(corr.columns)), corr.columns)  
```

```python
plot_corr(df)
```

Ao analisarmos as correlações com a tabela e com o gráfico acima, percebemos que as únicas variáveis que possuem uma correlação significativa são as variáveis "Peso" e "Altura".

Com um coeficiente de Pearson de 0,79, significa que possuem uma correlação forte e que dependendo do objetivo e da precisão necessária para o nosso projeto poderiamos criar um modelo preditivo para, com base em uma variável, prever a outra.

<b>Vamos agora enxergar a distribuição desses dados de peso e altura em um gráfico, juntamente com a reta aproximada que representa a distribuição:</b>



```python
sns.jointplot("Peso","Altura", df, kind = 'reg');
```

Cada ponto no gráfico mostra um atleta com seu respectivo peso (eixo x) e altura (eixo y), é até natural pensarmos que exista uma relação positiva entre as variáveis, ou seja, quanto maior o peso do atleta mais alto ele tende a ser.
